## Concept notes
### Objective function
An objective function measures what you are trying to maximize or minimize.

Technical note: Examples include maximizing expected return, minimizing variance, or minimizing prediction error.

Finance example: A portfolio optimizer might minimize volatility for a given return target.

### Constraints
Constraints limit what choices are allowed.

Technical note: Constraints can include weights summing to one, no shorting, leverage limits, or turnover caps.

Finance example: A portfolio with no-shorting cannot assign negative weights.

### Trade-offs
Optimization is usually about balancing competing goals.

Technical note: Higher return may come with higher risk, and cleaner predictions may require more model complexity.

Finance example: A high-return portfolio may be unacceptable if it violates risk or concentration limits.

## Code Lab 1: Brute-force portfolio search

```python
import numpy as np

cov = np.array([[0.04, 0.01], [0.01, 0.02]])
candidates = np.linspace(0, 1, 11)
results = []

for w in candidates:
    weights = np.array([w, 1 - w])
    variance = weights @ cov @ weights
    results.append((w, variance))

best = min(results, key=lambda x: x[1])
print("Best weight in asset 1:", best[0])
print("Minimum variance:", round(best[1], 4))
```

```python
TASK = 'Minimize portfolio variance with long-only weights that sum to one.'
print('Lab task:', TASK)

from pathlib import Path
import numpy as np
import pandas as pd
from scipy.optimize import minimize

market = pd.read_csv(Path("curriculum/datasets/real_market_prices.csv"), parse_dates=["date"])
prices = market.pivot(index="date", columns="symbol", values="close").dropna()
returns = prices.pct_change().dropna().iloc[:, :4]
cov = returns.cov().values
n = cov.shape[0]
x0 = np.ones(n) / n
bounds = [(0, 1)] * n
cons = ({"type": "eq", "fun": lambda w: w.sum() - 1},)
obj = lambda w: float(w @ cov @ w)
res = minimize(obj, x0=x0, bounds=bounds, constraints=cons)
print("success:", res.success)
print("weights:", np.round(res.x, 4))
print("variance:", round(float(res.fun), 8))
```
