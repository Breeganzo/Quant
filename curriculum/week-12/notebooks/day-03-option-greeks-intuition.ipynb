## Code Lab 1: Topic-specific implementation

```python
import numpy as np

k = 100
s = np.linspace(70, 130, 13)
call = np.maximum(s - k, 0)
put = np.maximum(k - s, 0)
eps = 1.0
s0 = 100
delta = (max(s0 + eps - k, 0) - max(s0 - eps - k, 0)) / (2 * eps)
print("call:", call)
print("put:", put)
print("delta near strike:", round(float(delta), 4))
```
