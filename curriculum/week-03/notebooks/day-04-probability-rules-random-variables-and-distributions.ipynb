## Concept notes
### Events and probabilities
An event is something that may happen, and probability measures how likely it is.

Technical note: Probabilities satisfy basic rules such as non-negativity, normalization, and additivity under disjoint events.

Finance example: A positive-return day or a default event can be treated as events in a probability model.

### Random variables
A random variable assigns a number to each uncertain outcome.

Technical note: Returns, losses, and payoffs are often modeled as random variables in finance.

Finance example: A one-day portfolio return is a random variable because it is not known in advance.

### Distribution thinking
A distribution describes the range of possible values and how likely they are.

Technical note: Good decisions in finance depend on the whole distribution, not just the average.

Finance example: A strategy with the same average return can still be worse if it has much fatter downside tails.

## Code Lab 1: Discrete payoff distribution

```python
import numpy as np

outcomes = np.array([3, 1, -2])
probabilities = np.array([0.2, 0.5, 0.3])
print("Expected payoff:", np.sum(outcomes * probabilities))
```

## Code Lab 2: Simulate repeated draws

```python
rng = np.random.default_rng(0)
samples = rng.choice(outcomes, size=1000, p=probabilities)
print("Sample mean:", round(samples.mean(), 4))
print("Sample std:", round(samples.std(), 4))
```

## Practice recap
- Why does a quant care about distributions and not only means?
  Suggested answer: Because risk, drawdown, and extreme outcomes live in the distribution's spread and tails, not only in its average.

## Real Market Data Lab (Useful From Day 1)

This section uses a local CSV snapshot of real market prices so the notebook remains reproducible.

Dataset: `curriculum/datasets/real_market_prices.csv`
Symbols: SPY, QQQ, TLT, GLD

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

data_path = Path("curriculum/datasets/real_market_prices.csv")
market = pd.read_csv(data_path, parse_dates=["date"]).sort_values(["date", "symbol"])

print("Rows:", len(market))
print("Date range:", market["date"].min().date(), "to", market["date"].max().date())
print("Symbols:", sorted(market["symbol"].unique()))
print(market.head())

prices = market.pivot(index="date", columns="symbol", values="close").dropna()
returns = prices.pct_change().dropna()

summary = pd.DataFrame(
    {
        "ann_return": returns.mean() * 252,
        "ann_vol": returns.std() * (252 ** 0.5),
        "max_drawdown": ((1 + returns).cumprod().div((1 + returns).cumprod().cummax()) - 1).min(),
    }
).sort_index()
print("\nRisk/return summary:")
print(summary.round(4))

cum = (1 + returns).cumprod()
cum.plot(figsize=(10, 5), title="Cumulative growth (base 1.0)")
plt.ylabel("Growth of $1")
plt.tight_layout()
plt.show()
```
