## Concept notes
### Arrays as financial objects
An array is a structured collection of numbers. In quant work those numbers might be prices, returns, weights, signals, or volumes.

Technical note: NumPy arrays support vectorized operations, broadcasting, and matrix-style computations that are much faster and clearer than many explicit Python loops.

Finance example: A daily return vector across 100 assets is naturally stored as one array.

### Vectorization
Vectorization means applying the same operation to many values at once.

Technical note: Instead of looping through each asset return manually, array operations act on the whole structure in one expression.

Finance example: Computing portfolio returns over many days becomes one matrix multiplication rather than repeated manual sums.

### Shape and dimension discipline
You need to know whether data are one-dimensional or two-dimensional before using them correctly.

Technical note: Shape mismatches are a common source of quant coding bugs, especially in ML pipelines and portfolio analytics.

Finance example: A weight vector and a return matrix must align properly for portfolio calculations to be correct.

## Code Lab 1: Vectorized returns

```python
import numpy as np

prices = np.array([100, 102, 101, 104, 106], dtype=float)
returns = prices[1:] / prices[:-1] - 1
print("Prices:", prices)
print("Returns:", np.round(returns, 4))
```

## Code Lab 2: Matrix of returns and portfolio weights

```python
returns_matrix = np.array(
    [
        [0.01, -0.02, 0.03],
        [0.02, 0.01, -0.01],
        [-0.01, 0.00, 0.02],
    ]
)
weights = np.array([0.5, 0.3, 0.2])
portfolio_returns = returns_matrix @ weights
print(np.round(portfolio_returns, 4))
```

## Code Lab 3: Loop versus vectorized calculation

```python
loop_result = []
for row in returns_matrix:
    loop_result.append(sum(row * weights))

print("Loop result:", np.round(loop_result, 4))
print("Vectorized result:", np.round(portfolio_returns, 4))
```
