## Concept notes
### APIs and JSON
An API is a way for one program to request data from another. JSON is a common text format used to return structured data.

Technical note: Understanding nested structures, keys, and parsing is enough to begin inspecting API outputs in finance workflows.

Finance example: A market-data endpoint may return timestamps, prices, and metadata in JSON format before you convert them into a table.

### Conditional probability
Conditional probability asks how likely an event is given that another event has happened.

Technical note: P(A|B) measures the probability of A under the condition that B is known.

Finance example: The probability of loss given a volatility spike is a conditional question.

### Bayes intuition
Bayes updates beliefs when new evidence arrives, but it does not ignore the prior baseline.

Technical note: Posterior belief combines prior belief with the likelihood of the new evidence.

Finance example: One strong signal should update your confidence, but not erase the base rate of noisy market behavior.

## Code Lab 2: Simple Bayes update

```python
prior = 0.10
likelihood_if_true = 0.70
likelihood_if_false = 0.20

posterior = (likelihood_if_true * prior) / (
    likelihood_if_true * prior + likelihood_if_false * (1 - prior)
)
print("Posterior probability:", round(posterior, 4))
```

## Practice recap
